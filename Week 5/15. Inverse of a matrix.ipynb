```python
def matrix():
    col=int(input("Enter the number of columns in matrix: "))
    row=int(input("Enter the number of rows in matrix: "))
    lst=[[0 for i in range(col)] for j in range(row)]
    for i in range(row):
        for j in range(col):
            lst[i][j]=int(input("Enter the [{}][{}] element of the matrix: ".format(i+1,j+1)))
    return lst

def det(A):
    row_A=len(A)
    col_A=len(A[0])
    if row_A==col_A:
        if row_A==2:
            return (A[0][0]*A[1][1]-A[1][0]*A[0][1])
        else:
            d=0
            a=1
            for i in range(col_A): #iterating through 1st column
                B=[[0 for k in range(row_A-1)] for l in range(row_A-1)]
                for x in range(row_A-1):
                    for y in range(row_A-1):
                        b = y if y<i else y+1
                        B[x][y]=A[x+1][b]
                d=d+a*A[0][i]*(det(B))
                a=-a
            return d        
                    
    else:
        print(" Determinant of the matrix can't be determined..")

def transpose(A):
    row_A=len(A)
    col_A=len(A[0])
    B=[[0 for i in range(row_A)] for j in range(col_A)]
    for k in range(col_A):
        for l in range(row_A):
            B[k][l]=A[l][k]
    return B        
        
def inverse(A):
    row_A=len(A)
    col_A=len(A[0])
    C=[[0 for i in range(col_A)] for j in range(row_A)]
    if row_A==2:
        C[0][0]=A[1][1]
        C[0][1]=-A[0][1]
        C[1][0]=-A[1][0]
        C[1][1]=A[0][0]
        B=[[0 for i in range(col_A)] for j in range(row_A)]
        for i in range(row_A):
            for j in range(col_A):
                B[i][j]=(1/det(A))*C[i][j]

    else:
        C=[[0 for i in range(col_A)] for j in range(row_A)]#cofactor matrix
        for i in range(row_A):
            for j in range(col_A):
                M=[[0 for k in range(row_A-1)] for l in range(col_A-1)]#minor matrix after removing the ith row and the jth column.
                for x in range(row_A-1):
                    for y in range(col_A-1):
                        a = x if x<i else x+1
                        b = y if y<j else y+1
                        M[x][y]=A[a][b]
                c=pow(-1,i+j)
                C[i][j]=c*det(M)
        D=transpose(C)
        B=[[0 for i in range(col_A)] for j in range(row_A)]
        for i in range(row_A):
            for j in range(col_A):
                B[i][j]=(1/det(A))*D[i][j]
    return B


A=matrix()
B=inverse(A)

print("A is: ")
for row in A:
    print (row)
print("The inverse of A is: ")
for row in B:
    print (row)
```
